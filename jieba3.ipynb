```python
data = pd.read_csv(r"C:\Users\Salman\Desktop\ML1\ML1\smalltrain.csv")
```

```python
chineseColumn1 = data.title1_zh.to_string()
```

We need to figure out what is the difference between cut_all=False/True in the below line. Link: https://github.com/fxsjy/jieba

```python
chineseSegGen = jieba.cut(chineseColumn1, cut_all=False)
```

```python
#Converting generator type to list 
chineseSegList = list(chineseSegGen)
```

```python
print(chineseSegList) #run from the begining once again
```

```python
#removing blank spaces
while ' ' in chineseSegList:
    chineseSegList.remove(' ')
```

```python
print(chineseSegList)
```

```python
chineseSegList.index('11')
```

```python
cvectorizer = CountVectorizer()
```

```python
y= cvectorizer.fit_transform(chineseSegList)
```

```python
cfeatures = cvectorizer.get_feature_names()
```

```python
print(cfeatures)
```

```python
print('Vocab count :',len(cfeatures))

print('Number of words:',len(chineseSegList))
```

```python
#converting sparse matrix to an Array
arrayMatrix = y.toarray()
```

```python
#converting array to a list
arrayList=arrayMatrix.tolist()
```

```python
#printing all the elements of counter vector
print(arrayList)
```
